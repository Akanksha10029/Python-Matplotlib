# Matplotlib: Axes Labels, Legend, Grid

```python
import matplotlib.pyplot as plt
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]

plt.plot(days, max_t, label="max",color="green")
plt.plot(days, min_t, label="min",color="blue")
plt.plot(days, avg_t, label="average",color="red")
```

**Axes labels and chart title**

```python
# set axes labels and chart title
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label="max",color="green")
plt.plot(days, min_t, label="min",color="purple")
plt.plot(days, avg_t, label="average",color="orange")
```

**Legend**

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label="max",color="green")
plt.plot(days, min_t, label="min",color="purple")
plt.plot(days, avg_t, label="average",color="orange")
plt.legend(loc="best",fontsize="large",shadow=True)
plt.grid(True)
```
