#WAP in Python to implenent the following:
**You are provided with a dataset containing climate data for a specific region over a period of time. The dataset includes information on temperature, rainfall, and humidity for each month. Your task is to create a complex plot with multiple subplots to visualize these climate parameters simultaneously.**
1.   months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
2.   temperature = [10, 12, 15, 18, 22, 26, 28, 27, 24, 20, 15, 12]
3.   rainfall = [40, 38, 42, 45, 55, 65, 75, 70, 60, 50, 45, 42]
4.   humidity = [60, 58, 62, 65, 68, 72, 76, 74, 70, 68, 62, 58]
**Create three subplots arranged in a grid (e.g., 2x2) to visualize temperature, rainfall, and humidity over the months.**
Customize each subplot by:
Adding labels and titles.
Using different colors and styles for the lines or bars.
Customizing the x-axis and y-axis labels.
Adding legends to distinguish between the climate parameters.
Highlighting specific data points or trends (e.g., maximum temperature or highest rainfall month).
**Apply appropriate data visualization techniques, such as line plots or bar charts, to represent the data effectively.**

```python
import matplotlib.pyplot as plt

# Data
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
temperature = [10, 12, 15, 18, 22, 26, 28, 27, 24, 20, 15, 12]
rainfall = [40, 38, 42, 45, 55, 65, 75, 70, 60, 50, 45, 42]
humidity = [60, 58, 62, 65, 68, 72, 76, 74, 70, 68, 62, 58]

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(18, 12))

# Subplot 1: Temperature
axs[0, 0].plot(months, temperature, marker='o', color='r', linestyle='-', label='Temperature')
axs[0, 0].set_title('Temperature Over Months')
axs[0, 0].set_xlabel('Months')
axs[0, 0].set_ylabel('Temperature (°C)')
axs[0, 0].legend()
axs[0, 0].grid(True)

# Subplot 2: Rainfall
axs[0, 1].bar(months, rainfall, color='b', label='Rainfall')
axs[0, 1].set_title('Rainfall Over Months')
axs[0, 1].set_xlabel('Months')
axs[0, 1].set_ylabel('Rainfall (mm)')
axs[0, 1].legend()
axs[0, 1].grid(True)

# Subplot 3: Humidity
axs[1, 0].plot(months, humidity, marker='s', color='g', linestyle='--', label='Humidity')
axs[1, 0].set_title('Humidity Over Months')
axs[1, 0].set_xlabel('Months')
axs[1, 0].set_ylabel('Humidity (%)')
axs[1, 0].legend()
axs[1, 0].grid(True)

# Subplot 4: Highlighting Maximum Temperature
max_temp_month = months[temperature.index(max(temperature))]
axs[1, 1].plot(months, temperature, marker='o', color='r', linestyle='-', label='Temperature')
axs[1, 1].set_title('Temperature with Max Highlight')
axs[1, 1].set_xlabel('Months')
axs[1, 1].set_ylabel('Temperature (°C)')
axs[1, 1].legend()
axs[1, 1].grid(True)
axs[1, 1].annotate(f'Max Temp ({max(temperature)}°C)', xy=(max_temp_month, max(temperature)), xytext=(3, 20),
            textcoords='offset points', arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

# Adjust subplot layout
plt.tight_layout()

# Display the plot
plt.show()
```
