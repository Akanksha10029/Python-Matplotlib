#Matplotlib: Histogram

**In histograms, x axis contains a variable and y axis will be a frequency of that variable**

```python
import matplotlib.pyplot as plt
```

**We have a sample data of blood sugar level of different patients, we will try to plot number of patients by blood range and try to figure out how many patients are normal, pre-diabetic and diabetic**

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, rwidth=0.8) #In this example, with rwidth=0.8, there is some spacing between the bars, making it easier to distinguish the individual bins.
```

**bins parameter**

```python
#The number of bins we choose determines how many of these sugar ranges we want to create in your histogram.
'''The "bins" in this context are like buckets or groups that you use to organize the scores.
Each bin represents a specific range of scores.'''
plt.hist(blood_sugar,rwidth=0.5,bins=3)
```

**Histogram showing normal, pre-diabetic and diabetic patients distribution**

*   80-100: Normal
*   100-125: Pre-diabetic
*   125-150: Diabetic







```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95 '''95% of whole graph''', color='orange')
```

**bold texMutiple data samples in a histogramt**

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','yellow'],label=['men','women'])
plt.legend()
```

**function histtype=step,stepfilled,barstacked**

```python
''' function, the histtype parameter is used to specify the type of histogram you want to create in terms of how the bars or steps are drawn.
Setting histtype to "step" results in a histogram where the data is represented as a series of steps or lines rather than bars.'''
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,histtype='step',color='green')
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,histtype='stepfilled',color='red')
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,histtype='barstacked',color='brown')
```

**horizontal orientation**

```python
plt.xlabel("Number Of Patients")
plt.ylabel("Sugar Level")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, orientation='horizontal',color='magenta')
```
