```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
data = """Student_ID,Gender,Grade,Study_Hours,Test_Score
1,Male,9,2,70
2,Female,10,3,85
3,Male,9,1,65
4,Female,11,5,92
5,Male,10,3,78
6,Male,8,2,60
7,Female,9,2,75
8,Female,11,4,88
9,Male,10,3,79
10,Male,8,1,62
11,Female,11,5,90
12,Female,9,2,73
13,Male,10,3,81
14,Female,11,4,87
15,Male,8,2,68
16,Male,9,2,71
17,Female,10,3,83
18,Female,9,1,67
19,Male,11,4,89
20,Female,10,3,80"""

# Create DataFrame
data = [line.split(',') for line in data.split('\n')]
columns = data[0]
data = data[1:]
df = pd.DataFrame(data, columns=columns)

# Convert columns to appropriate data types
df = df.astype({'Student_ID': int, 'Grade': int, 'Study_Hours': int, 'Test_Score': int})

# Univariate Analysis
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.histplot(df['Test_Score'], kde=True)
plt.title('Test Score Distribution')

plt.subplot(1, 3, 2)
sns.countplot(x='Grade', data=df)
plt.title('Grade Counts')

plt.subplot(1, 3, 3)
sns.countplot(x='Gender', data=df)
plt.title('Gender Counts')

plt.tight_layout()
plt.show()

# Bivariate Analysis
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.scatterplot(x='Study_Hours', y='Test_Score', data=df)
plt.title('Study Hours vs Test Score')

plt.subplot(1, 3, 2)
sns.boxplot(x='Grade', y='Test_Score', data=df)
plt.title('Grade vs Test Score')

plt.subplot(1, 3, 3)
sns.pairplot(df, hue='Gender')
plt.title('Pair Plot')

plt.tight_layout()
plt.show()

# Categorical Data Visualization
plt.figure(figsize=(10, 5))

sns.barplot(x='Grade', y='Test_Score', data=df, hue='Gender')
plt.title('Test Score by Grade and Gender')

plt.show()

# Heatmap
plt.figure(figsize=(8, 6))
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')

plt.show()
```
