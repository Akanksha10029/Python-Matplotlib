#Matplotlib : Pie Charts

```python
import matplotlib.pyplot as plt
```

**Draw pie chart to track down Monthly Business Expenses**

```python
exp_vals = [800, 400, 600, 300, 750]
exp_labels = ["Rent", "Salaries", "Utilities", "Supplies", "Marketing"]
plt.pie(exp_vals,labels=exp_labels)
```

**Draw pie chart as perfect circle**

```python
plt.pie(exp_vals,labels=exp_labels,shadow=True)
plt.axis('equal')
plt.show()
```

**Show percentages for every pie. Specify radius to increase or decrease chart size text**

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%0.2f%%',radius=1.5)
plt.show()
```

**Explode: take out some part of the pie chart**

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, autopct='%0.2f%%',radius=1.05,explode=[0.1,0,0.2,0,0.1])
plt.show()
```

**boldcounterclock and angle properties text**

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, autopct='%0.2f%%',radius=1.05,explode=[0.1,0,0,0.1,0],counterclock=False, startangle=180)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, autopct='%0.2f%%',radius=1.05,explode=[0.1,0,0,0.1,0],counterclock=False, startangle=90)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, autopct='%0.2f%%',radius=1.05,explode=[0.1,0,0,0.1,0],counterclock=False, startangle=45)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, autopct='%0.2f%%',radius=1.05,explode=[0,0,0,0,0.1],counterclock=True)
'''By default counterclock is True as we can see in explode first one is always the heighest %age
and after this second pie it consider is in its anticlockwise direction ,i.e, it starts counting the pie from anticlockwise direction...
so if we make counterclock as False then it will start consider the pie count in clockwise direction'''
plt.show()
```
